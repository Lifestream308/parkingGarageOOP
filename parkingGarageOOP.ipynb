```python



class ParkingGarage():
    
    available_parking = [1, 2, 3]
    unpaid = []
    
    current_ticket = {
        "1" : "Paid",
        "2" : "Paid",
        "3" : "Paid",
    }
    
    def __init__(self, owner):
        self.owner = owner
        
    def nameOwner(self): 
        print(f"The owner of this parking garage is {self.owner}")

    def takeTicket(self): 
        spot = int(input(f"{self.available_parking} is available. What spot you want? "))
        if spot in self.available_parking:    
            self.available_parking.remove(spot)
            self.unpaid.append(spot)
            self.current_ticket[spot] = "Unpaid"
            print(f"You will be in parking spot {spot}")
        elif spot not in self.available_parking: 
            print("That is not an available spot")
        
    def payForParking(self):
        if len(self.unpaid) > 0: 
            ticket = int(input(f"Tickets {self.unpaid} have not been paid. Which ticket are you paying for? "))
            if ticket in self.unpaid: 
                cost = int(input("How much will you be paying? Enter amount greater than 0 "))
                if cost > 0: 
                    print("Thank you for paying. You have 15 minutes to leave.")
                    self.unpaid.remove(ticket)
                    self.available_parking.append(ticket)
                    self.current_ticket[ticket] = "Paid"
        elif len(self.unpaid) == 0: 
            print("All tickets are currently paid for.")
        
    def leave(self):
        if len(self.unpaid) > 0: 
            receipt = int(input(f"Spot {self.unpaid} has not been paid for. What spot did you park in? "))
            if receipt in self.available_parking: 
                print("That ticket has been paid. Have a nice day!")
            elif receipt in self.unpaid: 
                print("That ticket is still unpaid for. Go Pay for your parking spot.")
        elif len(self.unpaid) == 0: 
            print("Have a nice day!")
        
    def startProgram(self):
        answer = input("Park? Pay? or Leave? ")
        if answer.casefold() == "park":
            self.takeTicket()
            self.startProgram()
        if answer.casefold() == "pay":
            self.payForParking()
            self.startProgram()
        if answer.casefold() == "leave":
            self.leave()
            self.startProgram()
        
new_garage = ParkingGarage("Christian A.") 
new_garage.startProgram()
```
